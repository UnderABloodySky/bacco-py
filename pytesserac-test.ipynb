

# Qué estamos haciendo?

*   imread :: Se carga la imagen
*   cvtColor :: Convierte la imagen a escala de grises
*   medianBlur :: Aplica un filtro de mediana para reducir el ruido
*   adaptiveThreshold :: Aplica una umbralización adaptativa
*   bitwise_not :: Invierte los colores de la imagen (negativo)
*   image_to_string :: Pasa la imagen a través de Pytesseract



```python
import cv2
import pytesseract

pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract' # Esto especifica la ruta al ejecutable de tesseract


image_path = path+"example.jpg"
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 3)
threshold_img = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
inverted_img = cv2.bitwise_not(threshold_img)
text = pytesseract.image_to_string(inverted_img)

print("Texto extraído:", text)
```
