```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Define la fuente y el tamaño del texto
font = ImageFont.load_default()

# Crea una imagen en blanco
image = Image.new("RGB", (200, 100), (255, 255, 255))
draw = ImageDraw.Draw(image)

# Genera texto sintético y lo dibuja en la imagen
text = "Texto de ejemplo"
draw.text((10, 10), text, font=font, fill=(0, 0, 0))

# Guarda la imagen generada
image_path = "synthetic_text_image.png"
image.save(image_path)

# Muestra la imagen utilizando matplotlib
img = Image.open(image_path)
plt.imshow(img)
plt.axis('off')  # Desactiva los ejes
plt.show()
```
